```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("..//image//001Bulbasaur.png")
plt.imshow(img)
```

```python
k = np.ones((50,50))
img_erode = cv2.erode(img, k)
img_open = cv2.dilate(img_erode, k)
plt.imshow(img_open)
```

```python
k = np.ones((15,15))
img_dilate = cv2.dilate(img, k)
img_close = cv2.erode(img_dilate, k)
plt.imshow(img_close)
```

```python
img_copy = img.copy()
for i in range(4):
    img_copy = cv2.pyrDown(img_copy)
    plt.imshow(img_copy)
    plt.show()
```
