```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("..//image//001Bulbasaur.png",0)
plt.imshow(img)
```

# BINARY

```python
threahold, thres = cv2.threshold(img,150,100,cv2.THRESH_BINARY)
plt.imshow(thres)
```

threshold, thres_inva = cv2.threshold(img, 150, 100, cv2.THRESH_BINARY_INV)
plt.imshow(thres_inv)

```python
threshold, thres_trunc = cv2.threshold(img, 150, 255, cv2.THRESH_TRUNC)
plt.imshow(thres_trunc)
```

```python
threshold, thresh_zero = cv2.threshold(img, 0, 150, cv2.THRESH_TOZERO)
plt.imshow(thresh_zero)
```

```python
threshold, thres_zero_inv = cv2.threshold(img, 150, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thres_zero_inv)
```

```python
thresh = cv2.adaptiveThreshold(img, 100, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 8)
plt.imshow(thresh)
```

```python
thresh = cv2.adaptiveThreshold(img, 11, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 8)
plt.imshow(thresh)
```

```python
origin, thresh = cv2.threshold(img, 150, 450, cv2.THRESH_TOZERO+cv2.THRESH_OTSU)
plt.imshow(thresh)
```
