```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("..//image//line.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
can_img = cv2.Canny(gray,300,300)
plt.imshow(can_img)
```

# Hough Line detection

```python
lines = cv2.HoughLines(can_img, 2, np.pi/180, 200,0)
print(lines.shape)

img = cv2.imread("..//image//line.jpg")

for line in lines:
    row, theta = line[0]
    
    cos = np.cos(theta)
    sin = np.sin(theta)
    
    x0 = cos*row
    y0 = sin*row
    
    x1 = int(x0 + 1000*(-sin))
    y1 = int(y0 + 1000*(cos))
    x2 = int(x0 - 1000*(-sin))
    y2 = int(y0 - 1000*(cos))
    
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),1)
    
plt.imshow(img)
```

# Circle Detection

```python
img = cv2.imread("..//image//circles.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
can_img = cv2.Canny(gray,150,150,apertureSize = 3)
plt.imshow(can_img)
```
